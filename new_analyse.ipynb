```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SheinHaul_channel_video_statistics.csv', on_bad_lines='skip')
df
```

```python
df_grouped = df.groupby('channelId')['SHvideoCommentCount'].sum()
df_grouped
```

```python
plt.figure(figsize=(16,8))
df['SHvideoDate'].value_counts().sort_index().plot(kind='line')
plt.show()
```

```python
top_10_dates = df['SHvideoDate'].value_counts().nlargest(10)
top_10_dates
```

```python
numeric_columns = df.dtypes[((df.dtypes=="float64")|(df.dtypes=="int64"))].index.values.tolist()
df[numeric_columns].hist(figsize=[21,8], layout=(3, 10))

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,  figsize=(21, 8))
df['channelTitle'].value_counts().plot(kind="bar").set_title('channelTitle')
plt.xticks(rotation=75)
fig.tight_layout()
fig.show()
```
